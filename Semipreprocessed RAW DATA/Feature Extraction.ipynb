```python
 # Import builder as main
# Implements a sliding window approach with given window size and overlap

def slidingWindow(key, size, overlap=0.5):
  
    if size > len(key.index):
        raise Exception('*Error* size of window can not be larger than input size..')
    if type(overlap) != type(0.0) or type(size) != type(0):
        raise Exception('*Error* type(overlap) must be float and type(size) must be int')
    if overlap > size:
       raise Exception('*Error* overlap is larger than size...')

    print('### Number of lines in frame: ', len(key.index), ' ###')
    
# Creating chunks
#windows = []
    windows = pd.DataFrame()
    steps = int(size * (1-overlap))
    print(steps)
    print(size)
    
# For all possible chunks in input
    for index in range(0, len(key.index) - steps, steps):

# Extract features in each sliding window

        extracted_features = extract_features(key.iloc[index: index + size], column_id="label")
        extracted_features.shape
        extracted_features.head

        windows=windows.append(extracted_features)
                
    windows.head
    print('### Number of windows: ', len(windows), ' ###')
    return windows
```
